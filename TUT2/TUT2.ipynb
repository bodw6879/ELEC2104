## Stuff To Remember
 
 - Conductor: no band gap; semiconductor: small band gap (~1 eV); insulator: large band gap (> ~9 eV)
 - $n$: number (or concentration - context dependent) of electrons in the conduction band
 - $p$: number of missing electrons (holes) in the valence band
 - $E_g = E_C - E_V$: band-gap energy - energy required to move one electron from the valence band to the conduction band
 - $n_i$: the carrier concentration/density for some material typically given in units per cm^3. The i subscript indicates the material is intrinsic (ie has not been doped)
 
For an intrinsic semiconductor, $n = p$. For an n-type semiconductor (doped with an element with 5 valence electrons), $n > p$. For a p-type semiconductor(doped with an element with 3 valence electrons), $p < n$.

Law of mass action: $np = n^2_i$, even with doping.

Summary for doped semiconductors:

 - Semiconductors can be doped to change their conductivity
 - In doped semiconductors, one of the two (electrons or holes) is the majority carrier
 - Drift current is the result of an electric field applied on the semiconductor
 - Diffusion current can be generated by varying the doping concentration and/or the doping type across a region of semiconductor
 - Both holes and electrons can contribute to drift and diffusion
 - The total current through a semiconductor can usually be approximated as just the majority-carrier drift current

## Question 1

Intrinsic carrier concentration of germanium:

$$n^{Ge}_i = (1.66 \times 10^{15})T^{3/2}exp(\frac {-E_g } { 2k_BT})$$

where $E_g = 0.66 eV$ is the band-gap energy, and $k_B = 1.38 \times 10^{-23} J/K$. Calculate $n_i$ at 300K and 600K. At high temperatures, how does the intrinsic carrier concentration vary with temperature?

```python
E_g = 0.66 * eV   # J
k_B = 1.38e-23   # J/K

def cc_germanium(T):
    return (1.66e+15) * T ** (3/2) * np.exp(-E_g / (2 * k_B * T))

# calculate carrier concentration for 300K and 600K:
print('For 300K: ' + str(cc_germanium(300)) + ', for 600K: ' + str(cc_germanium(600)))

# plot carrier concentration for some T
T = np.linspace(1, 1000, 999)
fig = plt.figure(figsize=(12, 8))
plt.plot(T, cc_germanium(T))
plt.title('Carrier Concentration vs Temperature', fontsize=16)
plt.xlabel('Temperature (K)', fontsize=16)
plt.ylabel('CC (cm^-3)', fontsize=16)

mpld3.plugins.connect(fig, mpld3.plugins.MousePosition())
mpld3.display()
```

## Question 2

If $n_i = 1.5 \times 10^{10} cm^{-3}$ at room temperature for silicon and a silicon wafer is doped with $N_A = 2 \times 10^{13} cm^{-3}$, determine the concentrations of free electrons, $n$, and free holes, $p$, in the wafer.

### Solution

Relevant slide:

$N_A$ is the acceptor impurity concentration (ie concentration of added atoms with 3 valence electrons). For large $N_A$, make the approximation $p = N_A$.
 
For p-type doping, the law of mass action implies $n = \frac {n^2_i} {N_A}$. We have $n_i = 1.5 \times 10^{10} cm^{-3}$, so that:

```python
n_i = 1.5e10   # per cm3
N_A = 2e13   # per cm3

p = N_A
n = n_i ** 2 / N_A

# print with scientific notation
print("n: %e, p: %e" % (n, p))
```

## Question 3

If $n_i = 1.5 \times 10^{10} cm^{-3}$ at room temperature for silicon and a silicon wafer is doped with $N_D = 3 \times 10^{16} cm^{-3}$, determine the concentrations of free electrons, $n$, and free holes, $p$, in the wafer.

### Solution

Same as above, but now we have n-type doping, where $N_D$ is the donor impurity concentration. Large enough $N_D$ means that approximately, $n = N_D$, and by the law of mass action we have: $$ p_n = \frac {n^2_i}  {N_D} $$
.

```python
n_i = 1.5e10   # per cm3
N_D = 3e16   # per cm3

n = N_D
p = n_i ** 2 / N_D

# print with scientific notation
print("n: %e, p: %e" % (n, p))
```

## Question 4

Calculate the conductivity of the wafer in Question 2, assuming $\mu_p = 480 cm^2V^{-1}s^{-1}$.

### Solution

Conductivity is given by: $$ \sigma = q(n\mu_n + p\mu_p) $$ <br>
where $\mu_n$ and $\mu_p$ are the electron mobility and hole mobility respectively. From Question 2 we have that $n = 1.125 \times 10^7$ and $p = 2 \times 10^{13}$. The wafer is positively doped, so the the number of electrons in the conductance band is going to be very small (several orders of magnitude) compared to the number of holes - so the approximation n = 0 can be made. Then we have that 
$$ \sigma = qp\mu_p $$.

```python
q = 1.60218e-19   # charge of 1 electron in Coulombs
p = 2e+13   # hole carrier concentration
hole_mobility = 480   # hole carrier mobility

conductivity = q * p * hole_mobility

print("Conductivity: %e (Ohm cm)^-1" % conductivity)
```

## Question 7

The concentration of electron holes at the right edge of a silicon bar of unit cross-sectional area is maintained at zero. Suppose the silicon bar is 0.1cm in length and that the concentration of holes varies linearly from the left edge of the silicon bar to the right edge of the bar. Given that the hole-diffusion current flowing through the bar is 1 mA, find the concentration of holes at the left edge of the bar. Use $D_P = 13 cm^2s^{-1}$ and the charge of an electron $q = -1.602 \times 10^{-19}$.

### Solution

The value $D_P$ is the hole diffusion constant that appears in $$ J_d = -qD_P\frac {dn} {dx} $$<br> $J_d$ is the diffusion current density, which is $ 1 mA cm^{-2} $. The solution to the differential equation is linear (the constant is 0):

$$ n = -\frac {1} {qD_p} x $$

```python
q = -1.60218e-19   # charge of 1 electron in Coulombs
D_P = 13   # hole diffusion constant

def n(x):
    return -(1 / (q * D_P)) * x

print("x = 0.1cm => n = %g (mA s) / (C cm)" % n(0.1))

# plot from 0 to 0.1
x = np.linspace(0, 0.1, 1000)

fig = plt.figure(figsize=(12, 8))
plt.plot(x, n(x))
plt.xlabel("distance along bar", fontsize=16)
plt.ylabel("hole carrier concentration", fontsize=16)
plt.title("(the x-axis is reversed here)", fontsize=16)

mpld3.plugins.connect(fig, mpld3.plugins.MousePosition())
mpld3.display()
```

## Question 8

The operation of semiconductor devices often has the situation in which excess carriers are injected into it, upsetting equilibrium conditions. Consider a p-type silicon bar into which electrons are injected at the left edge. It can be shown that recombination causes an exponential fall-off of electron concentration from the left edge. If $x = 0$ at the point of injection, the electron concentration density is given by $n_p(x) = n_p(0)exp(-x/L_n)$ where $L_n$ is the diffusion length for electrons. Assume the p-type silicon bar is 1 cm square in area and very long. Also assume the electron concentration at the point of injection ($x=0$) is $1014$ cm^-3. Calculate and plot the diffusion current betwteen $x = 0$ and $x = 0.1$ cm. Assume $D_n = 34 cm^{-2}s^{-1}$, the charge of an electron is $q = -1.602 \times 10^{-19}$ C, and $L_n = 0.01 cm$.

### Solution

We have: $$ n = n_0 e^{-\frac {x} {L_n}} $$ <br> 
$$ \frac {dn} {dx} = -\frac {n_0} {L_n} e^{-\frac {x} {L_n}} $$ <br>
$$ J_d = -qD_n\frac {dn} {dx} = q D_n \frac {n_0} {L_n} e^{-\frac {x} {L_n}} $$ <br>

Note $J_d$ is actually the diffusion current density, but the cross-sectional area is 1 so it's the same as the actual diffusion current.

```python
q = -1.60218e-19   # charge of 1 electron in Coulombs
D_n = 34   # hole diffusion constant
L_n = 0.01   # electron diffusion length in cm
n_0 = 1014   # cm^-3

def J_d(x):
    return q*D_n*n_0/L_n*np.exp(-x/L_n)

print("diffusion current at 0.1cm: %g A" % J_d(0.1))

x = np.linspace(0, 0.1, 1000)
fig = plt.figure(figsize=(12, 8))
plt.plot(x, J_d(x))
plt.title("Diffusion Current of Rod", fontsize=16)
plt.ylabel("diffusion current (A)", fontsize=16)
plt.xlabel("x (cm)", fontsize=16)

mpld3.plugins.connect(fig, mpld3.plugins.MousePosition())
mpld3.display()
```
