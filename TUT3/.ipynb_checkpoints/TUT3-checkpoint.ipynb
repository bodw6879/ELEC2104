## Formulas To Remember


 $$ W = \sqrt {\frac {2\epsilon_s} {q}(\frac {1} {N_A} + \frac {1} {N_D})(V_{bi} - v_F)} $$
 
 $$ W = \sqrt {\frac {2\epsilon_s} {q}(\frac {1} {N_A} + \frac {1} {N_D})(V_{bi} + v_R)} $$

 $$ V_{bi} = V_Tln(\frac {N_A N_D} {n_i^2}) $$
 
 $$ N_A x_p = N_D x_n $$
 
 $$ V_{bi} = \frac {1} {2} E_M W $$
 
 $$ C_j = \frac {\epsilon_s} {W_d} = \frac {\epsilon_s} {W_{d0}\sqrt{1 + \frac{v_R}{V_bi}}} = \frac {C_{j0}} {\sqrt {1 + \frac {v_R} {V_{bi}}}} $$

## Question 1

The two regions of a silicon diode are doped with concentrations $N_D = 10^{15} \:cm^{-3}$ and $N_A = 10^{18} \: cm^{-3}$. Calculate the minority carrier concentrations in each region. Use $n_i = 1.5\times10^{10} \:cm^{-3}$.

### Solution

The minority carriers are the carriers on the opposite sides - in the p region it's $n_p$ (concentration of conducting electrons outside the depletion region in the p-type silicon) and in the n region it's $p_n$ (concentration of holes outside the depletion region in the n-type silicon).

Use [Chapter 6](https://elearning.sydney.edu.au/bbcswebdav/pid-3132644-dt-content-rid-15555328_1/courses/2015_S2C_ELEC2104_ND/Lectures/Ch6new.pdf), slide 7. Since $n_n = N_D$ and $p_p = N_A$:

$$ V_{bi} = \frac {U_T} {q} ln(\frac {n_n p_p} {n_i^2}) \implies n_i^2 = n_n n_p e^{-V_{bi} q / U_T} $$ 

Also, using the law of mass action ($n_n p_n = n_i^2 = n_p p_p$):

$$ n_p = n_n e^{-V_{bi} q / U_T} $$
$$ p_n = p_p e^{-V_{bi} q / U_T} $$

Using the fact that $ \frac {q V_{bi}} {U_T} = ln(\frac {n_n n_p} {n_i^2}) $, we get:

$$ n_p = \frac {n_i^2} {p_p} $$


$$ p_n = \frac {n_i^2} {n_n} $$

.

```python
N_D = 1e+15   # cm^-3
N_A = 1e+18   # cm^-3
n_i = 1.5e+10   # cm^-3

n_n = N_D
p_p = N_A

n_p = n_i ** 2 / (p_p)
p_n = n_i ** 2 / (n_n)

print('n_p: %g cm^-3, n_n: %g cm^-3' % (n_p, p_n))
```

## Question 2

Calculate the barrier voltage across the depletion region of a silicon diode for a temperature of T = 300 K. You can use $V_T = 0.026 \:V$ and $n_i = 1.5 \times 10^{10} \: cm^{-3}$.

### Solution

Barrier voltage is another term for the built-in voltage, $V_{bi}$, which is given by the equation:

$$ V_{bi} = V_T ln( \frac {N_A N_D} {n_i^2} ) $$

We need to know $N_A$ and $N_D$. Using the values from question 1:

```python
N_D = 1e+15   # cm^-3
N_A = 1e+18   # cm^-3
n_i = 1.5e+10   # cm^-3

V_T = 0.026   # V

V_bi = V_T * np.log( N_A * N_D / ( n_i ** 2 ) )

print('V_bi: %g V' %V_bi)
```

## Question 4

Study the temperature dependence of the built-in voltage of a silicon diode which is doped with concentrations of
$N_D = 5 \times 10^{17} \: cm^{-3}$ and $N_A = 4 \times 10^{16} \: cm^{-3}$. Use $V = (k_BT/q)ln(N_AN_D/n_i^2)$ and $n_i = 5.2 \times 10^{15} T^{3/2} exp(-E_g/(2k_BT))$, where $E_g$ is the bandgap energy for silicon and is 1.12 eV. You can use $k_B = 8.6186 \times 10^{-5} \: eV/K $. Find the built-in voltage at 250K, 300K, and 350K. Does the built-in voltage go up or down with temperature?

```python
# convert to SI where necessary
N_D = 5e+17   # cm^-3
N_A = 4e+16   # cm^-3
E_G = 1.12 * 1.602e-19   # bandgap in J
k_B = 1.3806e-23   # Boltzmann constant in SI (J / K)
q = 1.602e-19   # fundamental charge

# T in K
def V_bi(T):
    n_i = 5.2e+15 * T ** (3/2) * np.exp( -E_G / (2 * k_B * T) )
    return (k_B * T / q) * np.log(N_A * N_D / (n_i ** 2))

print('V_bi at 250K: %g V; at 300K: %g V; at 350K: %g V' % (V_bi(250), V_bi(300), V_bi(350)))

# plot it
T = np.linspace(20, 400, 1000, dtype=np.float64)
V = V_bi(T)

fig = plt.figure(figsize=(12, 8))
plt.title('Built-in Voltage vs Temperature', fontsize=14)
plt.xlabel('Temperature (K)', fontsize=14)
plt.ylabel('Built-in Voltage (V)', fontsize=14)
plt.plot(T, V, '--r')

mpld3.plugins.connect(fig, mpld3.plugins.MousePosition())
plt.show()
```

## Question 5

A mobile phone VCO requires a variable capacitance with the characteristic shown in the figure below. Using $\epsilon_{Si} = \epsilon_{t}\epsilon_0$, where $\epsilon_t = 11.7$, $\epsilon_0=8.854\times10^{-14} \: F/cm$ and $q = 1.602\times10^{-19} \: C$, determine the value of $\frac {N_A N_D} {N_A + N_D}$ for a varactor diode that will do the job. If we choose $N_A = 2\times10^{18} \: cm^{-3}$, what should $N_D$ be?

<img src='fig1.png'>

### Solution

Modelling a reverse biased diode as a variable capacitor: let $ R = \frac {N_A N_D} {N_A + N_D} $. We have that:


$$C_j = \frac {\epsilon_S} {W_d} = \sqrt{\frac{\epsilon_S^2} {\frac {2\epsilon_s} {q}(\frac {1} {N_A} + \frac {1} {N_D})(V_{bi} + v_R)}} = \sqrt {\frac {q\epsilon_S} {2} \frac {N_A N_D} {N_A + N_D} \frac {1} {V_{bi} + v_R}} = \sqrt {\frac {q\epsilon_S} {2} R \frac {1} {V_{bi} + v_R}}$$

There are two unknowns here - $R$ and $V_{bi}$. Since we have two points, we can solve for these - rearranging the above:

$$ R = (V_{bi} + v_R)\frac {2} {q\epsilon_S} C_j^2 $$
.

```python
# convert to SI units, except for length which we convert to cm - note that the 'f' prefix is the SI prefix for femto (10^-15)
C1 = 1.3 * 1e-15 / ( 1e-4 ** 2 )   # F / cm^2
V1 = -1.5   # V
C2 = 2.2 * 1e-15 / ( 1e-4 ** 2 )   # F / cm^2
V2 = -0.5   # V

print('C1: %g F / cm^2; V1: %g V; C2: %g F / cm^2; V2: %g V;' % (C1, V2, C2, V2))
print('C1^2: %g; C2^2: %g;' % (C1 ** 2, C2 ** 2))
```

So for the points we have:

$$ R = (V_{bi} - 1.5)\frac {2} {q\epsilon_S} 1.69\times10^{-14}, \:\:\: R = (V_{bi} - 0.5)\frac {2} {q\epsilon_S} 4.84\times10^{-14} $$

Further, we can rearrange $R$ to get:

$$ N_A R + N_D R = N_A N_D \implies N_D = \frac {N_A R} {N_A - R} $$

.

```python
q = -1.602e-19   # C
epsilon_S = 11.7 * 8.854e-14   # F / cm

coef_R1 = 1
coef_V_bi1 = - (2 / ( q * epsilon_S)) * C1 ** 2
const_1 = -1.5 * (2 / (q * epsilon_S)) * C1 ** 2

coef_R2 = 1
coef_V_bi2 = - (2 / ( q * epsilon_S)) * C2 ** 2
const_2 = -0.5 * (2 / ( q * epsilon_S)) * C2 ** 2

# solve the linear system
A = np.array([[coef_R1, coef_V_bi1], [coef_R2, coef_V_bi2]], dtype=np.float64)
b = np.array([const_1, const_2], dtype=np.float64)
R, V_bi = np.linalg.solve(A, b)

print('The linear system is: \n')
print(coef_R1, '* R +', coef_V_bi1, '* V_bi =', const_1)
print(coef_R2, '* R +', coef_V_bi2, '* V_bi =', const_2)

print('\nWith solution:\n ')

print('R: %g, V_bi: %g V \n' % (R, V_bi))

print('If N_A = 2e+18, then N_D = %g' % (2e+18 * R / (2e+18 - R)))
```

```python
# Plot C_j against v_R for good measure

# returns F / cm^2
def C_j(v_R):
    return np.sqrt( (q * epsilon_S / 2) * R / (V_bi + v_R) )

v_R = np.linspace(-0.01, -3, 1000)
y = C_j(v_R)

# convert y to fF / micrometer^2
y = y * 1e+15 / (1e+4 ** 2)

fig = plt.figure(figsize=(12, 8))
plt.title('Reverse Biased Diode Capacitance vs Voltage', fontsize=14)
plt.xlabel('Voltage (V)', fontsize=14)
# plt.ylabel('Capacitance (fF / um^2)', fontsize=14)
plt.plot(v_R, y, '--r')

# plot two specific points of interest, to check that the parameters R and V_bi are correctly fitted
points_x = np.array([-0.5, -1.5])
points_y = C_j(points_x) * 1e+15 / (1e+4 ** 2)
plt.plot(points_x, points_y, 'ro')

mpld3.plugins.connect(fig, mpld3.plugins.MousePosition())
plt.show()
```

## Question 6

A voltage of 0.65 V is applied in the forward direction to a silicon diode which has $I_S = 2\times10^{-14} \: A$ at room temperature (300K). Determine the diode current at room temperature.

### Solution

The diode current-voltage equation is: 

$$ I = I_S (e^{V/V_T} - 1) $$

Remember that thermal voltage is given by $V_T = \frac {k_B T} {q}$.

```python
k_B = 1.3806488e-23   # J / K
q = 1.602e-19   # C
T= 300   # K
I_S = 2e-14   # A
V = 0.65   # applied voltage, V

V_T = k_B * T / q

I = I_S * (np.exp(V / V_T) - 1)

print('Current through forward-biased diode: %g A' % I)
```

## Question 7

A silicon diode with $I_S = 2\times10^{-14} \: A$ has a diode current of 2 mA in the forward direction. What is the diode voltage?

### Solution

Assuming room temperature as above, we have that:

$$ \frac {I} {I_S} = e^{V/V_T} - 1 \implies V = V_T ln(\frac {I} {I_S} + 1) $$

.

```python
k_B = 1.3806488e-23   # J / K
q = 1.602e-19   # C
T= 300   # K
I_S = 2e-14   # A
I = 2e-3   # resultant current, A

V_T = k_B * T / q   # thermal voltage, V

V = V_T * np.log(I / I_S + 1)

print('voltage across forward-biased diode: %g V' % V)
```

## Question 8

For the curcuit shown, diode D1 has $I_S = 2\times10^{-15} \: A$. Calculate the current $I_X$ and the voltage across diode D1 when $V_X = 1\:V$. Use $V_T = 0.026 \: V$. Note that the circuit equation is non-linear. 

<img src='fig2.png'>

### Solution

Note the voltage configuration foward-biases the diode. By Ohm's law, $V_X = RI_X + V_D$. By the FB diode current-voltage equation, $I_X = I_S(e^{V_D/V_T} - 1) $ implies $ V_D = V_T ln(\frac {I} {I_S} + 1)$. So we have:

$$ V_X = RI_X + V_Tln(\frac {I_X} {I_S} + 1) $$

To 'solve' this equation we need to find the value of $I_X$ that makes this equation true. It happens that this very difficult analytically, so define:

$$ z(I_X) = RI_X + V_Tln(\frac {I_X} {I_S} + 1) - V_X $$

and use a numerical optimisation procedure to find the value of I_X that gives $z(x) = 0$.
<br>
<br>


```python
import scipy.optimize as opt
V_X = 1   # V
V_T = 0.026   # V
I_S = 2e-15   # A
R = 2e+3   #  Ohms

def z(I_X):
    return R*I_X + V_T * np.log(I_X / I_S + 1) - V_X

# try using the Newton-Raphson method
I_X_optimal = opt.newton(z, 1)

# we can find V_D using
V_D = V_T * np.log(I_X_optimal / I_S + 1)

print('I_X: %g A, V_D: %g V' % (I_X_optimal, V_D))

# alternatively, we can solve for I_X_optimal graphically:
# we need to first guess the interval for I_X - say 0 to 0.001 A
I_X = np.linspace(0, 0.001, 1000)   # A

fig = plt.figure(figsize=(12, 8))
plt.plot(I_X, z(I_X))
plt.title('')

mpld3.plugins.connect(fig, mpld3.plugins.MousePosition())
plt.show()
```
